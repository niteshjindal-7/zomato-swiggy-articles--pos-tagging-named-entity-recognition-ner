```python
'''Create function to perform POS tagging, NER, and syntantic dependency parsing ''' 
def spacy_ner(data, articlewithdocs_name):

    def spacy_ops(corpus):
        corpus_annotated = []
        corpus_lemmatized = []
        corpus_pos = []
        for i in range(len(corpus)):
            doc1 = nlp(corpus[i])  
            out1 = [(token.text, token.lemma_, token.pos_, token.tag_) for token in doc1]
            corpus_lemmatized.append([token.lemma_ for token in doc1])
            corpus_pos.append([token.pos_ for token in doc1])
            corpus_annotated.append(out1)
        return corpus_lemmatized, corpus_pos, corpus_annotated
    
    reviews = data['Article_content']
    reviews = list(reviews)
    corpus_lemmatized, corpus_pos, corpus_annotated = spacy_ops(reviews)  
    labels1 = ['word', 'lemma', 'POS', 'xpos_tag']
    
    '''POS Tagging ''' 
    pos = pd.DataFrame(corpus_annotated[0], columns = labels1)
    for i in range(1, len(corpus_annotated)):
        pos_df1 = pd.DataFrame(corpus_annotated[i], columns = labels1)
        pos = pos.append(pos_df1)
        print("\nGet Part-of-speech tagged words in -"+articlewithdocs_name)
        display(pos)
    
    '''Named Entity Recognition'''
    labels11 = ['Token', 'Entity']
    ner1 = pd.DataFrame()
    for i in range(0, len(reviews)):
        doc_1 = nlp(reviews[i])
        ner_1 = pd.DataFrame([(X.text, X.label_) for X in doc_1.ents], columns = labels11)
        ner_1['doc_num'] = i
        ner1 = ner1.append(ner_1)
        print("\nGet named entity recognition in -"+articlewithdocs_name)
        display(ner1)
    
    
    ''' Syntantic Dependency Parsing'''
    sdp =pd.DataFrame()
    for i in range(0,len(reviews)):
        doc = nlp(reviews[i])
        for token in doc:
            sdp1 =pd.DataFrame([token.text, token.lemma_, token.pos_, token.dep_]).transpose()
            sdp = sdp.append(sdp1)
            #print("\nsyntactic dependency in -"+articlewithdocs_name)
            #display(sdp)
    return pos, ner1, sdp
```

```python
loop = (Zomato_data_during_incidence.dropna(), Zomato_data_after_incidence.dropna(), 
        Zomato_data_before_incidence.dropna(), Swiggy_data_during_incidence.dropna(), 
        Swiggy_data_after_incidence.dropna(), Swiggy_data_before_incidence.dropna())

for i in range(0, len(loop)):
    j = ["zoma_POS&NER_During", "zoma_POS&NER_After", "zoma_POS&NER_Before", 
         "swgy_POS&NER_During", "swgy_POS&NER_After", "swgy_POS&NER_Before"]
    print("\n\n")
    display("Get Part-of-speech tagged words, named entity recognition, syntactic dependency in -"+str(j[i]))
    results = spacy_ner(loop[i], articlewithdocs_name=str(j[i]))
    pos = results[0]
    ner1 = results[1]
    sdp = results[2]
    #if i<3:
    with pd.ExcelWriter("C:/Users/AG89382/AppData/Local/Programs/Python/Python37/deepenv/Swiggy-Zomato-Food-Aggregators-Analysis/Articles_Data/pos_&_ner_output/" + j[i] + ".xlsx") as writer:
        pos.to_excel(writer, sheet_name= 'POS_Tagging')  
        ner1.to_excel(writer, sheet_name= 'NamedEntityRecognition')
        sdp.to_excel(writer, sheet_name= 'SyntanticDependency')
    #else:
     #   with pd.ExcelWriter("C:/Users/XXXXX/AppData/Local/Programs/Python/Python37/deepenv/Swiggy-Zomato-Food-Aggregators-Analysis/Articles_Data/" + j[i] + ".xlsx") as writer:
      #      pos.to_excel(writer, sheet_name= 'POS_Tagging')  
       #     ner1.to_excel(writer, sheet_name= 'NamedEntityRecognition')
       #     sdp.to_excel(writer, sheet_name= 'SyntanticDependency')
print("--- %s seconds ---" % (time.time() - start_time))
```
